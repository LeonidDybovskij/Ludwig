```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('ks-projects.csv', delimiter=',')
```

```python
data
```

```python
print(data.columns.tolist())
```

```python
from ludwig.api import LudwigModel
```

```python
model_definition = {
    "input_features": [
        {"name": "category ", "type": "category"},
        {"name": "main_category ", "type": "category"},
        {"name": "currency ", "type": "category"},
        {"name": "deadline ", "type": "date"},
        {"name": "goal ", "type": "numerical"},
        {"name": "launched ", "type": "date"},
        {"name": "pledged ", "type": "numerical"},
        {"name": "backers ", "type": "numerical"},
        {"name": "country ", "type": "category"},
        {"name": "usd pledged ", "type": "numerical"},
    ],
    "output_features": [
        {"name": "state ", "type": "category"}
    ]
}
```

```python
training_dataframe = data[:323057]
test_dataframe = data[323057:]
```

```python
model = LudwigModel(model_definition)
```

```python
train_stats = model.train(training_dataframe)
```

```python
predictions = model.predict(test_dataframe)
```

```python
model.save("Kickstarter")
```
